```python
import pandas as pd
```

```python
from sklearn.datasets import load_wine
```

```python
wine = load_wine()
```

```python
df = pd.DataFrame(wine.data, columns = wine.feature_names)
```

```python
df["target"] = wine.target
```

```python
df
```

```python
df["target"].value_counts()
```

```python
df.isna().sum()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize = (20,12))
df.boxplot()
plt.show()
```

```python
df.info()
```

```python
import seaborn as sns
corr = df.corr()
plt.figure(figsize = (12,6))
sns.heatmap(corr,annot = True)
plt.show()
```

```python
X = df.drop(columns = "target")
X
```

```python
y = df["target"]
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, stratify=y)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X_train_norm = scaler.fit_transform(X_train)
X_train_norm
```

```python
X_test_norm = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
log_model = LogisticRegression()
```

```python
log_model.fit(X_train_norm, y_train)
```

```python
log_model.score(X_train_norm, y_train)
```

```python
log_model.score(X_test_norm, y_test)
```

```python
log_model.predict_log_proba(X_test_norm)
```

```python
log_model.predict_proba(X_test_norm)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
tree_model = DecisionTreeClassifier()
```

```python
tree_model.fit(X_train_norm, y_train)
```

```python
from sklearn.tree import plot_tree
```

```python
plt.figure(figsize = (10,6))
plot_tree(tree_model, filled = True)
plt.show()
```

```python
tree_model.score(X_train_norm, y_train)
```

```python
tree_model.score(X_test_norm, y_test)
```

```python
from sklearn.svm import SVC
```

```python
svc_model = SVC()
```

```python
svc_model.fit(X_train_norm, y_train)
```

```python
svc_model.score(X_train_norm, y_train)
```

```python
svc_model.score(X_test_norm, y_test)
```

#### Hard Voting - Voting Classifer

```python
from sklearn.ensemble import VotingClassifier
```

```python
model1 = LogisticRegression()
model2 = DecisionTreeClassifier()
model3 = SVC(probability=False)
```

```python
voting_clf = VotingClassifier(estimators=[
    ("lr", model1),
    ("dt", model2),
    ("svc", model3)
], voting="hard")
```

```python
voting_clf.fit(X_train_norm, y_train)
```

```python
voting_clf.fit_transform(X_train_norm, y_train)
```

```python
voting_clf.fit_transform(X_test_norm, y_test)
```

```python
voting_clf.score(X_train_norm, y_train)
```

```python
voting_clf.score(X_test_norm, y_test)
```

#### Soft Voting - Voting Classifier

```python
model1 = LogisticRegression()
model2 = DecisionTreeClassifier()
model3 = SVC(probability=True)
```

```python
voting_clf = VotingClassifier(estimators=[
    ('lr', model1), 
    ('dt', model2), 
    ('svc', model3)
], voting='soft')
```

```python
voting_clf.fit(X_train_norm, y_train)
```

```python
voting_clf.fit_transform(X_train_norm, y_train)
```

```python
voting_clf.fit_transform(X_test_norm, y_test)
```

```python
voting_clf.score(X_train_norm, y_train)
```

```python
voting_clf.score(X_test_norm, y_test)
```

#### Bagginng (sampling with replacement) : set parameter "bootstrap = True"

```python
from sklearn.ensemble import BaggingClassifier
```

```python
bag_clf = BaggingClassifier(
    DecisionTreeClassifier(), n_estimators=500, max_samples = 100, bootstrap = True, n_jobs = -1
)
```

```python
bag_clf.fit(X_train_norm, y_train)
```

```python
bag_clf.score(X_train_norm, y_train)
```

```python
bag_clf.score(X_test_norm, y_test)
```

#### Pasting - sampling without replacement - bootstrap = False

bag_clf = BaggingClassifier(
    DecisionTreeClassifier(), n_estimators=500, max_samples = 100, bootstrap = False, n_jobs = -1
)

```python
bag_clf.fit(X_train_norm, y_train)
```

```python
bag_clf.score(X_train_norm, y_train)
```

```python
bag_clf.score(X_test_norm, y_test)
```

#### Out-of-bag evaluation

```python
bag_clf = BaggingClassifier(
    DecisionTreeClassifier(), n_estimators=500, max_samples = 100, bootstrap = True, n_jobs = -1, oob_score=True
)
```

```python
bag_clf.fit(X_train_norm, y_train)
```

```python
bag_clf.oob_score_
```

```python
bag_clf.score(X_test_norm, y_test)
```

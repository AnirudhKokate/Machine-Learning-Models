```python
df = pd.read_csv("dataset/Student_Performance.csv")
```

```python
def Yes_No(vals):
    if (vals == "Yes"):
        return 1
    elif (vals == "No"):
        return 0
df["Extracurricular Activities"] = df["Extracurricular Activities"].apply(Yes_No)
```

```python
df.head()
```

## Visualization Methods

 Pair Plot (Scatterplot Matrix)

```python
import seaborn as sns
```

```python
sns.pairplot(df)
```

Correlation Heatmap (Very Informative)

```python
import matplotlib.pyplot as plt

# Compute correlation matrix
corr_matrix = df.corr()

# Create heatmap
#plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Heatmap')
plt.show()
```

Regression Plots for All Features in One Loop

```python
for feature in df.columns[:-1]:  # Exclude the target column
    sns.lmplot(x=feature, y="Performance Index", data=df)
    plt.title(feature + " vs Target")  # Normal string concatenation
    plt.show()
```

## Training and testing

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
X = df.drop(columns = "Performance Index")
X.head()
```

```python
y = df[["Performance Index"]]
y.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25)
```

```python
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_train, y_train)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
y_predict = model.predict(X_test)
y_predict = pd.DataFrame(y_predict)
y_predict.rename({0: "Predicted"},axis = 1, inplace = True)
y_predict.head()
```

```python
y_test
```

```python
y_test.reset_index(inplace = True)
```

```python
y_test.drop(columns = "index", inplace = True)
```

```python
comp = pd.concat([y_predict, y_test], axis = 1)
comp
```

```python
model.score(X_test, y_test)
```

```python
metrics.mean_squared_error(y_test, y_predict)
```

```python
metrics.mean_absolute_error(y_test, y_predict)
```

```python
metrics.mean_squared_error(y_predict, y_test)
```

```python
metrics.mean_absolute_error(y_predict, y_test)
```

```python
import matplotlib.pyplot as plt

plt.scatter(y_test, y_predict, color='blue', alpha=0.2)
plt.xlabel("Actual Values (y_test)")
plt.ylabel("Predicted Values (y_predict)")
plt.title("Actual vs. Predicted Performance Index")
plt.axline((0, 0), slope=1, color='red', linestyle="--", label="Perfect Fit")  # Optional
plt.plot()
plt.legend()
plt.show()
```

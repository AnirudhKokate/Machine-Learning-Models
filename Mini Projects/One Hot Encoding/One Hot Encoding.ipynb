```python
import pandas as pd
```

```python
df = pd.read_csv("homeprices.csv")
df.head()
```

```python
df.shape
```

```python
dummies = pd.get_dummies(df["town"])
dummies
```

```python
dummies = dummies.astype("int64")
dummies
```

```python
merged = pd.concat([df,dummies], axis = 1)
merged
```

```python
merged = merged.drop(columns = "town")
merged
```

Dropping one of the dummy variables as The third column (e.g., west windsor) is completely determined by the first two columns.
If using tree-based models (e.g., Decision Trees, Random Forests), dropping a column is not necessary because these models are not affected by multicollinearity.

```python
final = merged.drop(columns = "west windsor")
final
```

Simple Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
X = final.drop(columns = "price")
X
```

```python
y = final[["price"]]
y
```

```python
model = LinearRegression()
```

```python
model.fit(X,y)
```

```python
model.score(X,y)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
y_predict = model.predict(X)
y_predict = pd.DataFrame(y_predict)
y_predict
```

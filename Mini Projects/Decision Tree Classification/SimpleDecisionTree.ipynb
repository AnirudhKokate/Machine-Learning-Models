```python
import pandas as pd
```

```python
df = pd.read_csv("salaries.csv")
df.head()
```

```python
X = df.drop(columns = "salary_more_then_100k")
X.head()
```

```python
y = df[["salary_more_then_100k"]]
y.head()
```

## Label Encoding 
Label Encoding does not inherently know or respect the logical order of categories. It simply assigns integers based on the alphabetical order (or lexicographical order) of the categories by default.

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le_company = LabelEncoder()
le_job = LabelEncoder()
le_degree = LabelEncoder()
```

```python
X["company_n"] = le_company.fit_transform(X["company"])
X["job_n"] = le_company.fit_transform(X["job"])
X["degree_n"] = le_company.fit_transform(X["degree"])
```

```python
X.head()
```

```python
X = X.drop(columns = ["company", "job", "degree"])
X.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
model = DecisionTreeClassifier()
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
y_predict = model.predict(X_test)
y_predict
```

```python
metrics.accuracy_score(y_predict, y_test)
```

### Visualizing the Tree

```python
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree
```

```python
plt.figure(figsize = (10,6))
plot_tree(model, filled = True)
plt.show()
```
